## 1. Eigenface for face recognition. (40 pts)
In this assignment you will implement the Eigenface method for recognizing human faces. You
will use face images from The Yale Face Database B, where there are 64 images under different
lighting conditions per each of 10 distinct subjects, 640 face images in total.

### (a) (2 pts) 

Download The Face Dataset and unzip faces.zip , You will find a folder called images

and test (validation) set split respectively, each line gives an image path and the correspond-

```python
def format_plot(ax, xlab = "", ylab = "", title = "", legend = False, xticks = None, yticks = None):
    ax.set_xlabel(xlab,  fontsize = 14)
    ax.set_ylabel(ylab,  fontsize = 14)
    ax.set_title(title,  fontsize = 16, fontweight = "bold")
    if legend:
        ax.legend(fontsize = 14)
    if xticks is not None:
        ax.set_xticks(xticks)
    if yticks is not None:
        ax.set_yticks(yticks)
```

### (b) (2 pts) 
Load the training set into a matrix X: there are 540 training images in total, each has
50 × 50 pixels that need to be concatenated into a 2500-dimensional vector. So the size of X
should be 540 × 2500, where each row is a flattened face image. Pick a face image from X and
display that image in grayscale. Do the same thing for the test set. The size of matrix X test for
the test set should be 100 × 2500.

```python
import numpy as np
import imageio.v2 as imageio
from matplotlib import pylab as plt
import matplotlib.cm as cm
%matplotlib inline

train_labels, train_data = [], []
for line in open('./faces/train.txt'):
    im = imageio.imread(line.strip().split()[0])
    train_data.append(im.reshape(2500, ))
    train_labels.append(line.strip().split()[1])
train_data, train_labels = np.array(train_data, dtype = float), np.array(train_labels, dtype = int)

test_labels, test_data = [], []
for line in open('./faces/test.txt'):
    im = imageio.imread(line.strip().split()[0])
    test_data.append(im.reshape(2500, ))
    test_labels.append(line.strip().split()[1])
test_data, test_labels = np.array(test_data, dtype = float), np.array(test_labels, dtype = int)

print("Train Data Shape: ", train_data.shape, "\tTrain Labels Shape: ", train_labels.shape)
print("Test Data Shape: ", test_data.shape, "\tTest Labels Shape: ", test_labels.shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].imshow(train_data[10, :].reshape(50, 50), cmap = cm.Greys_r)
ax[1].imshow(test_data[10, :].reshape(50, 50), cmap = cm.Greys_r)

format_plot(ax[0], xticks = [], yticks = [], title = "10th Train Image")
format_plot(ax[1], xticks = [], yticks = [], title = "10th Test Image")

plt.show()
```

### (c) (3 pts) 
Average Face. Compute the average face µ from the whole training set by summing up
every row in X then dividing by the number of faces. Display the average face as a grayscale image.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))

avg_face = train_data.sum(0) / len(train_labels)

ax.imshow(avg_face.reshape(50, 50), cmap = cm.Greys_r)
format_plot(ax, xticks = [], yticks = [], title = "Average Face")

plt.show()
```

### (d) (3 pts) 
Mean Subtraction. Subtract average face $µ$ from every row in $X$. That is, $x_i := x_i −
µ$, where $x_i$ is the i-th row of $X$. Pick a face image after mean subtraction from the new $X$
and display that image in grayscale. Do the same thing for the test set $X$ test using the pre-
computed average face $µ$ in (c).

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

train_data_avg = train_data - avg_face
test_data_avg = test_data - avg_face

ax[0].imshow(train_data_avg[10, :].reshape(50, 50), cmap = cm.Greys_r)
ax[1].imshow(test_data_avg[10, :].reshape(50, 50), cmap = cm.Greys_r)

format_plot(ax[0], xticks = [], yticks = [], title = "10th Train Image Minus Avg Face")
format_plot(ax[1], xticks = [], yticks = [], title = "10th Test Image Minus Avg Face")

plt.show()
```

### (e) (10 pts) 
Eigenface. Perform eigendecomposition on $X^TX = VΛV^T$ to get eigenvectors $V^T$,
where each row of $V^T$ has the same dimension as the face image. We refer to $v_i$, the i-th
row of $V^T$, as i-th eigenface. Display the first 10 eigenfaces as 10 images in grayscale.

```python
from numpy.linalg import eig

fig, ax = plt.subplots(2, 5, figsize = (14, 6))

_, V = eig(np.cov(train_data_avg.T))
efaces = np.real(V).T.reshape(2500, 50, 50)

for i in range(2):
    for j in range(5):
        idx = j + 5 * i
        ax[i, j].imshow(efaces[idx, :, :], cmap = cm.Greys_r)
        format_plot(ax[i, j], xticks = [], yticks = [], title = "Eigenface n" + str(idx))

plt.show()
```

### (f) (10 pts) 
Eigenface Feature. The top $r$ eigenfaces $V^T[: r, :] = \{v_1, v_2, ..., v_r \}^T$ span an $r$-dimensional
linear subspace of the original image space called face space, whose origin is the average face
$µ$, and whose axes are the eigenfaces $\{v_1, v_2, ..., v_r \}$. 

Therefore, using the top $r$ eigenfaces $\{v_1, v_2, ..., v_r \}$, we can represent a 2500-dimensional face image $z$ as an $r$-dimensional feature vector $f: f = V^T[:r, :] z = [v_1 , v_2 , ..., v_r]^Tz$. Write a function to generate $r$-dimensional feature matrix $F$ and $F_{test}$  for training images $X$ and test images $X_{test}$, respectively 

(to get $F$, multiply $X$ to the transpose of first $r$ rows of $V^T$ , $F$ should have same number of rows as $X$ and $r$ columns; similarly for $X_{test}$ ).

```python
def get_eigenface_features(X, V, r):
    return X @ V.reshape(2500, -1)[:r, :].T
```

```python
F = get_eigenface_features(train_data, efaces, 10)
F_test = get_eigenface_features(test_data, efaces, 10)

print("F shape: ", F.shape, "\tF_test shape: ", F_test.shape)
```

### (g) (10 pts) 
Face Recognition. For this problem, you are welcome to use libraries such as scikit
learn to perform logistic regression. Extract training and test features for r = 10. Train a
Logistic Regression model using F and test on F test . Report the classification accuracy on the
test set. Plot the classification accuracy on the test set as a function of r when r = 1, 2, . . . , 200.
Use “one-vs-rest” logistic regression, where a classifier is trained for each possible output
label. Each classifier is trained on faces with that label as positive data and all faces with
other labels as negative data. sklearn calls this the “ovr” mode.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class = "ovr", max_iter = 5000)
model.fit(F, train_labels)

log_regr_score = model.score(F_test, test_labels)
print ("Accuracy score for r = 10 is: " + str(log_regr_score))
```

```python
score_arr = []

for r in range(1, 201):    
    F = get_eigenface_features(train_data, efaces, r)
    F_test = get_eigenface_features(test_data, efaces, r)
    model.fit(F, train_labels)
    log_regr_score = model.score(F_test, test_labels)
    score_arr.append(log_regr_score)

fig, ax = plt.subplots(1, 1, figsize = (12, 6))

ax.plot(range(1, 201), score_arr)
format_plot(ax, xlab = "Top 'r' EigenFaces", ylab = "Score", title = "Classification Accuracy on the Test Set")

plt.show()
```

### (b) (3 pts) 
Recall that EM learns the parameter $θ$ of a Gaussian mixture model $P_θ(x, z)$ over a dataset $D = \{x^{(i)} | i = 1, 2, ...n\}$ by performing the E-step and the M-step for $t = 0, 1, 2, ...$ . 

We repeat the E-step and and M-step until convergence. In the E-step, for each $x^{(i)} ∈ D$, we compute a vector of probabilities $P_{θ_t}(z = k | x)$ for the event that each $x^{(i)}$ originates from a cluster $k$ given the current set of parameters $θ_t$.

Write the expression for $P_{θ_t}(z = k | x)$, which is the posterior of each data point $x^{(i)}$. Recall
that by Bayes’ rule,

$$
P_{θ_t}(z = k | x) = {P_{θ_t}(z = k, x) \over P_{θ_t}(x)} 
= {P_{θ_t}(z = k, x) \over \sum_{l=1}^{K} P_{θ_t}(x|z = l)P_{θ_t}(z = l)} 
$$


Note that we have seen this formula in class. We are asking you to write it down and try to
understand and it before implementing it in part (e).

#### Answer: 

$$
P_{θ_t}(z = k | x) = {P_θ(z = k, x) \over P_θ(x)} 
                   = {P_θ(x|z = k)P_θ(z = k) \over \sum_{l=1}^K P_θ(x|z = l)P_θ(z = l)}
$$

### (c) (5 pts) 
In the M-step, we compute new parameters $θ_{t+1}$. Our goal is to find $µ_k$, $Σ_k$ and $φ_k$ that optimize

$$
max_θ \bigg( \sum_{k=1}^{K}\sum_{x ∈ D} P_{θ_t}(z_k | x) \log P_θ(x | z_k) +
       \sum_{k=1}^{K}\sum_{x ∈ D} P_{θ_t}(z_k | x) \log P_θ(z_k)\bigg)
$$

Write down the formula for $µ_k$ , $Σ_k$, and for the parameters $φ$ at the M-step (we have also seen
these formulas in class).

#### Answer:

If we define the number of training targets with class $k$ as $n_k = |\{i : y^{(i)} = k\}|$.

Then we can compute $µ_k$, $Σ_k$ and $φ_k$ as

$$
µ_k =  {\sum_{i=1}^{n}P(z = k | x^{(i)})x^{(i)} \over n_k}
$$

$$
Σ_k =  {\sum_{i=1}^{n}P(z = k | x^{(i)})(x^{(i)} - µ_k)(x^{(i)} - µ_k)^T \over n_k}
$$

$$
φ_k = {n_k \over n}
$$


### (d) (25 pts)
Implement and run the EM algorithm. Specifically

#### i. (10 pts) Implement the EM algorithm from scratch (e.g., in Python and numpy ).


```python
from scipy.stats import multivariate_normal as mv_normal

class GMM:
    def __init__(self, clusters = 2, rand_state = None):
        self.clusters = clusters
        self.rng = np.random.default_rng(seed = rand_state)
        
    def e_step(self, X, priors, means, covars):
        n, d = X.shape
        k = self.clusters
        T = np.ndarray((n, k))
        for i in range(k):
            T[:, i] = priors[i] * mv_normal.pdf(X, means[i], covars[i])
        return T / T.sum(1)[:, np.newaxis]

    def m_step(self, X, T):
        n, d = T.shape
        k = self.clusters
        
        tots = T.sum(0)
        priors = tots / n
        mus = (T.T @ X) / tots[:, np.newaxis]

        covars = np.ndarray(shape = (k, d, d))
        for i in range(k):
            X_centered = X - mus[i]
            covars[i] = (T[:, i] * X_centered.T) @ X_centered / tots[i]
        return priors, mus, covars
        
    def fit(self, X, conv_thresh = 1e-6):
        
        n, d = X.shape
        k = self.clusters
        
        Z = np.zeros((n, k))
        Z[np.arange(n), self.rng.choice(k, n)] = 1
        T = self.m_step(X, Z)

        has_converged = False
        mu_list = []

        while not has_converged:
            T0 = T
            Z = self.e_step(X, *T)
            T = self.m_step(X, Z)

            mu_list.append(T[1])        
            has_converged = abs(np.sum(np.square(T0[1] - T[1]))) < conv_thresh

        self.Z = Z
        self.mu_list = np.array(mu_list)
        
        return self
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

gmm = GMM()
gmm.fit(geyser_data)

clus1, clus2 = [], []
for i in range(0, 272):
    if gmm.Z[i][0] > gmm.Z[i][1]:
        clus1.append(geyser_data[i])
    else:
        clus2.append(geyser_data[i])

clus1, clus2 = np.array(clus1), np.array(clus2)
ax.scatter(clus1[: , 0], clus1[: , 1], s = 5, c = 'r', label = "cluster1")
ax.scatter(clus2[: , 0], clus2[: , 1], s = 5, c = 'b', label = "cluster2")
ax.scatter(*gmm.mu_list[-1].T, c = ['k', 'g'], s = 100)

format_plot(ax, xlab = "Geyser Eruptions", ylab = "Duration", title = "GMM Clustering Results", legend = True)

plt.show()
```

#### ii. (5 pts) Choose a termination criterion for when the algorithm stops repeating the E-step and the M-step. State your termination criterion and explain the reasoning behind it.


#### Termination Criterion: absolute difference between centers and previous centers < $10^{-6}$
To determine termination we defined a convergence threshold of $10^{-6}$.
At every step we compute the absolute difference between the new centers and the old ones, if the distance is smaller than the threshold we assume that the algorithm has converged and stop fitting.

We tuned the convergence threshold by trial and error. $10^{-6}$ resulted to be a value small enough to avoid converging at local optima, while being large enough to detect fast when it converged, ultimately avoiding to compute multiple centers really close to one another.


#### iii. (10 pts) Plot the trajectories of the two mean vectors ($µ_1$ and $µ_2$) in two dimensions as they change over the course of running EM. You might want to use a scatter plot for this.

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
for i, mus in enumerate(gmm.mu_list):
    mu1 = mus[0, :].T
    mu2 = mus[1, :].T
    
    if i < len(gmm.mu_list) - 1:
        mu1_next = gmm.mu_list[i+1][0, :].T
        mu2_next = gmm.mu_list[i+1][1, :].T
        
        ax.annotate(text = '', xy = (mu1[0], mu1[1]), xytext = (mu1_next[0], mu1_next[1]), 
                    arrowprops = dict(arrowstyle = '<-'))
        ax.annotate(text = '', xy = (mu2[0], mu2[1]), xytext = (mu2_next[0], mu2_next[1]), 
                    arrowprops = dict(arrowstyle = '<-', color = 'g'))

    
    ax.scatter(*mus.T, c = ['k', 'g'], s = 30)

ax.scatter(clus1[: , 0], clus1[: , 1], s = 5, c = 'r', label = "cluster1")
ax.scatter(clus2[: , 0], clus2[: , 1], s = 5, c = 'b', label = "cluster2")

format_plot(ax, xlab = "Geyser Eruptions", ylab = "Duration", title = "GMM Centers Evolution", legend = True)

plt.show()
```

### (e) (5 pts)
If you were to run K-means clustering instead of the EM algorithm you just implemented, do you think you will get different clusters? You are welcome to experiment with K-means clustering on the same dataset with $K = 2$. (The KNN library from scikit learn is a good way to try). Comment on why do you think the results will or will not change.

#### Answer :

The main difference between EM and K-means is that the former can get stuck at local optima. However, the small convergence threshold that we defined in (d) does a good job at avoiding that.

Therefore, we expected that the results should not largely differ. As the last figure shows, the centers identified by the two algorithms are very close to each other. However, the two algorithms slightly differ in their ability to sort the data in two clusters. The last figure clearly shows that k-means clusters as "blue" some points closer to the "red" cluster's mean, and viceversa.

Therefore, we found that overall EM did a better job in dividing the data in two distinct classes.


```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 2).fit(geyser_data, 2)

clus1, clus2 = [], []
for lab, pt in zip(kmeans.labels_, geyser_data):
    if label[lab] == 0:
        clus1.append(pt)
    else:
        clus2.append(pt)

clus1, clus2 = np.array(clus1), np.array(clus2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

ax.scatter(clus1[:, 0], clus1[:, 1], s = 5, c = 'r', label = "cluster1")
ax.scatter(clus2[:, 0], clus2[:, 1], s = 5, c = 'b', label = "cluster2")

ax.scatter(*gmm.mu_list[-1].T, c = 'k', s = 100, label = "GMM Centers")
ax.scatter(*kmeans.cluster_centers_.T, c = 'g', s = 100, label = "Kmeans Centers")

format_plot(ax, xlab = "Geyser Eruptions", ylab = "Duration", title = "GMM Clustering Results", legend = True)

plt.show()
```
